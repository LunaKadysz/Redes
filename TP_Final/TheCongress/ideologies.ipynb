El objetivo de este notebook es establecer para cada ao las ideologías representativas. Para esto nos quedamos con los partidos mayoritarios hasta tener cubierto por alguna ideología el ~80% de lxs diputadxs. 

```python
from procesador_votaciones import ProcesadorDeVotaciones
```

```python
procesador = ProcesadorDeVotaciones(1993, 2020, 'data')
congreso = procesador.procesar()
```

```python
def Sort_Tuple(tup):  
    return(sorted(tup, key = lambda x: x[1], reverse = True)) 

for year in congreso.votings:
    print(year)
    partidos = []
    for party in congreso.parties:
        gente = party.representatives.get(year)
        if gente:
            partidos.append((party.text, len(gente)))
    partidos_anos = Sort_Tuple(partidos)
    print(f'{partidos_anos} \n \n')
```
