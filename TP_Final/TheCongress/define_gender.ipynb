```python
procesador = ProcesadorDeVotaciones(1993, 2020, 'data')
congreso = procesador.procesar()
```

```python
import gender_guesser.detector as gender
d = gender.Detector()
d.get_gender(u"Uriel")
```

```python
gender_a_asignar = []
gender = {}
for representative in congreso.representatives:
    if representative.id:
        complete_name = representative.name
        first_name = complete_name.split(' ')[0]
        genero = d.get_gender(f'{first_name}')
        if not (genero == 'female' or genero == 'male'):
            second_name = complete_name.split(' ')[-1]
            new_genero = d.get_gender(f'{second_name}')
            if not (new_genero == 'female' or new_genero == 'male'):
                print(complete_name)
                gender_a_asignar.append(complete_name)
            else: 
                gender[complete_name] = new_genero
            
        else:
            gender[complete_name] = genero
    
```

```python
gender_a_mano = set(gender_a_asignar)
gender_a_mano.remove('Fabián Dulio')
gender_a_mano.remove('Javier Revnaldo')
gender_a_mano.remove('Gastón H.')
gender_a_mano.remove('Legislador 1')
gender_a_mano.remove('Legislador 2')
gender_a_mano.remove('Zulema Beatríz')
gender_a_mano.add('Zulema')
gender_a_mano.add('Beatriz')
gender_a_mano.remove('Analuz Ailén')
gender_a_mano.add('Ailén')
gender_a_mano.remove('Washington Jesus')
gender_a_mano.add('Jesus')
```

```python
genders = ['female', 'male', 'female', 'female','male', 'male', 'male', 'male', 'male', 'male', 'male', 'male', 'male', 'female', 'female','female', 'female','female','female','female', 'male', 'male', 'male','male','male','female', 'female']
```

```python
gender_dict = dict(zip(sorted(list(gender_a_mano)), genders))
```

```python
gender_dict
```

```python
def search_library(name_to_search, complete_name, dict_a_mano, dict_to_save):
    genero = d.get_gender(f'{name_to_search}')
    if genero == 'female' or genero =='male':
        dict_to_save[complete_name] = genero
        return True
    
    else:
        a_mano = dict_a_mano.get(name_to_search)
        if a_mano:
            dict_to_save[complete_name] = a_mano
            return True
        
        else:
            return None
```

```python
new_gender = {}
faltaron = []

for representative in congreso.representatives:
    if representative.id:
        if not 'Legislador' in representative.name:
            complete_name = representative.name
            found = search_library(complete_name, complete_name, gender_dict, new_gender)
            if not found :
                first_name = complete_name.split(' ')[0]
                found_first = search_library(first_name,complete_name,  gender_dict, new_gender)
                if not found_first:
                    second_name = complete_name.split(' ')[-1]
                    found_second = search_library(second_name,complete_name,  gender_dict, new_gender)
                    if not found_second:
                        faltaron.append(complete_name)    
```
