```python
from procesador_votaciones import ProcesadorDeVotaciones
```

```python
procesador = ProcesadorDeVotaciones(1993, 2020, 'data')
congreso = procesador.procesar()
```

```python
import gender_guesser.detector as gender
d = gender.Detector()
d.get_gender(u"Uriel")
```

```python
gender_a_asignar = []
gender = {}
for representative in congreso.representatives:
    if representative.id:
        complete_name = representative.name
        first_name = complete_name.split(' ')[0]
        genero = d.get_gender(f'{first_name}')
        if not (genero == 'female' or genero == 'male'):
            second_name = complete_name.split(' ')[-1]
            new_genero = d.get_gender(f'{second_name}')
            if not (new_genero == 'female' or new_genero == 'male'):
                print(complete_name)
                gender_a_asignar.append(complete_name)
            else: 
                gender[complete_name] = new_genero
            
        else:
            gender[complete_name] = genero
    
```

```python
gender_a_mano = set(gender_a_asignar)
gender_a_mano.remove('Fabián Dulio')
gender_a_mano.remove('Javier Revnaldo')
gender_a_mano.remove('Gastón H.')
gender_a_mano.remove('Legislador 1')
gender_a_mano.remove('Legislador 2')
gender_a_mano.remove('Zulema Beatríz')
gender_a_mano.add('Zulema')
gender_a_mano.add('Beatriz')
gender_a_mano.remove('Analuz Ailén')
gender_a_mano.add('Ailén')
gender_a_mano.remove('Washington Jesus')
gender_a_mano.add('Jesus')
```

```python
new_gender = {}
for representative in congreso.representatives:
    if representative.id:
        complete_name = representative.name
        first_name = complete_name.split(' ')[0]
        genero = d.get_gender(f'{first_name}')
        if not (genero == 'female' or genero == 'male'):
            otro_genero = gender_dict.get(first_name)
            if not (otro_genero == 'female' or otro_genero == 'male'):
                second_name = complete_name.split(' ')[-1]
                new_genero = d.get_gender(f'{second_name}')
                if not (new_genero == 'female' or new_genero == 'male'):
                    gender_a_asignar.append(complete_name)
                else:
                    new_gender[complete_name] = otro_genero
            
            else: 
                new_gender[complete_name] = new_genero
            
        else:
            new_gender[complete_name] = genero
    
```

```python
gender_dict.update(new_gender)
```

```python
import json
with open('configs/gender.json', 'w') as outfile:
    json.dump(gender_dict, outfile)
```
