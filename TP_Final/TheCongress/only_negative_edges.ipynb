```python
year = 2018
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
from procesador_votaciones import ProcesadorDeVotaciones
from heavy_graph_maker import HeavyGraphMaker
from disparity_filter import DisparityFilter
from new_disparity_filter import NewDisparityFilter
```

```python
procesador = ProcesadorDeVotaciones(year, year, 'data')
congreso = procesador.procesar()
```

```python
congreso.validate()
```

```python
maker = HeavyGraphMaker(congreso)
```

```python
grafo = maker.create_year_network(year, negative = True)
```

```python
x = grafo.get_sorted_edges_weights().values()
plt.hist(x, bins = 100)
plt.xlabel('Pesos de los ejes normalizados por cantidad total de votaciones')
plt.show()
```

```python
print(f' Ese anio se trataron {len(congreso.votings[year])} votaciones')
```

```python
fig, axs = plt.subplots(1, 1, figsize = (20, 20))
nodos = grafo.get_nodes()
pos = nx.spring_layout(grafo.graph)
ec = nx.draw_networkx_edges(grafo.graph, pos, alpha=0.2)
nc = nx.draw_networkx_nodes(grafo, pos, nodelist=nodos,
                        with_labels=False, node_size=10, cmap=plt.cm.jet)
axs.set_title(f'Grafo positivo del ao {year}')
axs.axis('off')
plt.savefig(f'../figs/negative_graphs/{year}.jpg')
plt.show()
```

```python
print(f' El grafo tiene {len(grafo.get_edges())} enlaces')
```

Lo cortamos segn el tamao de la CG. 

```python
copia = grafo.copy()
grafo_cortado = DisparityFilter(copia).cut_graph(GC_limit_size = 0.9)
```

```python
fig, axs = plt.subplots(1, 1, figsize = (20, 20))
nodos = grafo_cortado.get_nodes()
pos = nx.spring_layout(grafo_cortado.graph)
ec = nx.draw_networkx_edges(grafo_cortado.graph, pos, alpha=0.2)
nc = nx.draw_networkx_nodes(grafo_cortado, pos, nodelist=nodos,
                        with_labels=False, node_size=50, cmap=plt.cm.jet)
axs.set_title(f'Grafo cortado manteniendo la componente gigante')
axs.axis('off')

plt.savefig(f'../figs/negative_graphs/{year}_GC.jpg')
plt.show()
```

```python
x = grafo_cortado.get_sorted_edges_weights().values()
plt.hist(x, bins = 100)
plt.xlabel('Pesos de los ejes normalizados por cantidad total de votaciones')
plt.show()
```

Cortamos segn el valor de $\alpha$

```python
copia2 = grafo.copy()
grafo_cortado2 = DisparityFilter(copia2).cut_graph(alpha_limit = 0.002)
```

```python
fig, axs = plt.subplots(1, 1, figsize = (20, 20))
nodos = grafo_cortado.get_nodes()
pos = nx.spring_layout(grafo_cortado2.graph)
ec = nx.draw_networkx_edges(grafo_cortado2.graph, pos, alpha=0.2)
nc = nx.draw_networkx_nodes(grafo_cortado2, pos, nodelist=nodos,
                        with_labels=False, node_size=50, cmap=plt.cm.jet)
axs.set_title(f'Grafo cortado segn alpha')
axs.axis('off')
plt.savefig(f'../figs/negative_graphs/{year}_alpha.jpg')
plt.show()
```
