```python
procesador = ProcesadorDeVotaciones(1993, 2020, 'data')
congreso = procesador.procesar()
```

```python
for year,votes in congreso.votings.items():
    for vote in votes:
        votantes = vote.get_voters()
        for votante in votantes:
            if votante.id == None:
                print(f'En la votación: {vote.id} del {year} hay un None')
    
```

En el resto de votaciones hay Nones, pero que corresponden a votos AUSENTES. Léase no nos interesan. Por otra parte los nones se dan en momentos donde luego cambia une diputade. Pueden ser lapsus entre el recambio por muerte / renuncia / licencia. 

```python
def find_nones(voting_con_none, control, year):
    for voto in congreso.votings[year]:
        if voto.id == voting_con_none:
            votacion_rara = voto
        
    for voto in congreso.votings[year]:
        if voto.id == control:
            votacion_control = voto
            
    raros = votacion_rara.get_voters(ausentes = True)
    control = votacion_control.get_voters(ausentes = True)
    cambios = set(control).difference(raros)
    return cambios
        
    
```

```python
none_93 = find_nones(3622, 3623, 1993)
for none in none_93:
    print(none.name, none.last_name)
```
