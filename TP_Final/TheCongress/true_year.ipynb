No hay 257 casi en ningun año ja.Sólo en 2012.

Ahora lo hago sin diciembre

```python
cant_repre = []
for year in congress.votings.keys():
    repr_list = [repre for repre in congress.representatives if repre.was_in_year(year)]
    cant_repre.append(len(repr_list))
    print(f'{year}: {len(repr_list)}')
```

```python
import numpy as np
print(f'El promedio de diputadxs por año, sin diciembre, es {round(np.array(cant_repre).mean(), 2)} + - {round(np.array(cant_repre).std(), 2)}')
```

Opciones:
* Renuncias
* Muertes

* **Recambio en diciembre**

```python
cambios_en_dic = np.array(list(con_diciembre.values())) - np.array(cant_repre)
cambios_en_dic
```

Concluimos que los cambios en diciembre son significativos cada dos años. Por ahora vamos a correr la data sin el mes 12. Y luego agregar esas votaciones.

De paso, quiero ver si todas las votaciones tienen a 257 diputadxs.

```python
for year in range(1993, 2020):
    for ley in congress.votings[year]:
        diputa = len(ley.votos['AFIRMATIVO']) + len(ley.votos['NEGATIVO']) + len(ley.votos['ABSTENCION']) + len(ley.votos['AUSENTE'])
        if diputa != 257: 
            print(diputa, ley.text )
```

Tienen 257!

```python
from collections import defaultdict
votings_to_change = defaultdict(list)
for year in range(1993, 2020):
    old = congress.votings[year][0].get_voters(ausentes = True)
    for voting in congress.votings[year]:
        new = voting.get_voters(ausentes = True)
        diff=  set(new).difference(set(old))
        if len(diff) > 20:
            votings_to_change[year].append(voting)
            print(f'La ley: {voting.id} del mes {"diciembre" if voting.month == 12 else voting.month} de {year} tuvo {len(diff)} cambios')
        else:
            old = new 
```
